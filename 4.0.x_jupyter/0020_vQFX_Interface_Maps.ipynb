# 0020 Create AOS Interface Maps

Load AOS Class + get token/inventory from 0000

```python
import json
from pprint import pprint 
from Apstra import Apstra
%store -r inventory
print(inventory)
aos_server = Apstra(inventory)
```

## 0020A - Interface Map for Spine
![GUI](img/0020A.png "Create vQFX Interface Map")

```python
interface_maps_url = "/api/design/interface-maps"

f = open('json/0020A_design_interface-maps_AOS-12x10-Spine.json')
interface_map_spine = json.load(f)

resp = aos_server.http_post(interface_maps_url, interface_map_spine, expected=201)
```

![GUI](img/0020A_created.png "Create vQFX Interface Map")
